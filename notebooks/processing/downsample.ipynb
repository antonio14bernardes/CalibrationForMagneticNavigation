# Reduce positional resolution

Was only used for OctoMag ablation study.

```python
import pandas as pd
import os
import sys

cwd = os.getcwd()
parent_dir = os.path.dirname(cwd)
base_dir = os.path.dirname(parent_dir)
data_dir = base_dir + "/data/octomag_data/split_dataset" # Path to OctoMag data
src_dir = base_dir + "/src"
params_dir = cwd + "/params"

sys.path.insert(0, src_dir)
from data_analysis import plot_positions, data_processing, config
```

## How much to reduce train and validation?

```python
keep_percentage = 1 # percent. In the paper we used reductions to 50%, 20%, 5%, and 1%

resolution = 0.0016
```

## Load data

```python
# Load training and validation data
training_data = pd.read_pickle(data_dir + "/training_data.pkl")
validation_data = pd.read_pickle(data_dir + "/validation_data.pkl")
```

## Reduce data

Again, reduction done solely based on position. Fixed random seed used for reproducibility.

```python
get_rid = 100 - keep_percentage

# Reuse function for spliting the dataset
reduced_training_data, _, _ = data_processing.split_dataset_positional(training_data, keep_percentage/100, get_rid/100, 0, resolution=resolution, shuffle=True, random_state=config.SPLIT_SEED)
```

```python
# Visualize the reduced data
plot_positions(reduced_training_data, point_size=1.0, title="Reduced Training Data Positions")
```

```python
plot_positions(training_data, point_size=1.0, title="Full Training Data Positions")
```

```python
reduced_validation_data, _, _ = data_processing.split_dataset_positional(validation_data, keep_percentage/100, get_rid/100, 0, resolution=resolution, shuffle=True, random_state=config.SPLIT_SEED)
```

```python
plot_positions(reduced_validation_data, point_size=1.0, title=f"Reduced Validation Data Positions")
```

```python
plot_positions(validation_data, point_size=1.0, title="Full Validation Data Positions")
```

```python
reduced_training_data.to_pickle(data_dir + f"/training_data_{keep_percentage}.pkl")
reduced_validation_data.to_pickle(data_dir + f"/validation_data_{keep_percentage}.pkl")
```
